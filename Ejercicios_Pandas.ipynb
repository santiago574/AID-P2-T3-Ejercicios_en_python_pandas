# Ejercicio 1
#### Escribir un programa que pregunte al usuario por las ventas de un rango de años y muestre por pantalla una serie con los datos de las ventas indexada por los años, antes y después de aplicarles un descuento del 10%.

```python
import pandas as pd

inic = int(input('¿En que año inicia? '))
fin = int(input('¿En que año finaliza? '))
ven = {}
for i in range(inic, fin+1):
    ven[i] = float(input('¿Cuantás ventas hubó en el año ' + str(i) +'? '))
ven = pd.Series(ven)
print('Ventas\n', ven)
print('Ventas con descuento\n', ven-ven*10/100)
```

# Ejercicio 2
#### Escribir una función que reciba un diccionario con las notas de los alumnos en curso en un examen y devuelva una serie con la nota mínima, la máxima, media y la desviación típica.

```python
import pandas as pd

def dic_notas(notas):
    notas = pd.Series(notas)
    dic = pd.Series([notas.min(), notas.max(), notas.mean(), notas.std()], index=['Min', 'Max', 'Media', 'Desviación típica'])
    return dic

notas = {'Andress':9, 'Jesús':8.5, 'Erick':9.5, 'Giovanny': 10, 'Santiago': 9}
print(dic_notas(notas))
```

# Ejercicio 3
#### Escribir una función que reciba undiccionario con las notas de los alumnos en curso en un examen y devuelva una serie con las notas de los alumnos aprobados ordenadas de mayor a menor.

```python
import pandas as pd

def ap(notas):
    notas = pd.Series(notas)
    return notas[notas >= 6].sort_values(ascending=False)

notas = {'Andress':9, 'Jesús':5, 'Erick':9.5, 'Giovanny': 10, 'Santiago': 9}
print(ap(notas))
```

# Ejercicio 5
#### Escribir una función que reciba un DataFrame con el formato del ejercicio anterior, una lista de meses, y devuelva el balance (ventas -gastos) total en los meses indicados.

```python
import pandas as pd

dat = pd.DataFrame([['Enero', 30500, 22000],
                    ['Febrero', 35600, 23400], 
                    ['Marzo', 28300, 18100],
                    ['Abril', 33900, 20700]], 
                   columns= ['Mes','Ventas','Gasto'])
blankIndex=[''] * len(dat)
dat.index=blankIndex

def balance(datos, meses):
    dat['Bal'] = dat.Ventas - dat.Gasto
    return dat.set_index('Mes').loc[meses,'Bal'].sum()

print(balance(dat, ['Febrero','Abril']))
```

#### El fichero cotizacion.csv contiene las cotizaciones delas empresas del IBEX35 con las siguientes columnas: nombre(nombre de la empresa), Final(precio de la acción al cierre de bolsa), Máximo(precio máximo de la acción durante la jornada), Mínimo(precio mínimo de la acción durante la jornada), volumen(Volumen al cierre de bolsa), Efectivo(capitalización al cierre en miles de euros). Construir una función que construya un DataFrame a partir deun fichero con el formato anterior y devuelva otro DataFrame con el mínimo, el máximo y la media de dada columna.

```python
import pandas as pd

def resumen(fichero):
    df = pd.read_csv(fichero, sep=';', decimal=',', thousands='.', index_col=0)
    return pd.DataFrame([df.min(), df.max(), df.mean()], index=['Mínimo', 'Máximo', 'Media'])

resumen('/content/gdrive/MyDrive/Colab Notebooks/cotizacion 1.csv')
```

#### Los ficheros emisiones-2016.csv, emisiones-2017.csv, emisiones-2018.csvy emisiones-2019.csv, contienen datos sobre las emisiones contaminates en la ciudad de Madrid en los años 2016, 2017, 2018 y 2019 respectivamente. Escribir un programa con los siguientes requisitos:

```python
import pandas as pd
import numpy as np
import datetime as dt

import pandas as pd 

emisiones_2016 = pd.read_csv('/content/gdrive/MyDrive/Colab Notebooks/emisiones-2016 1.csv', sep = ';')
emisiones_2017 = pd.read_csv('/content/gdrive/MyDrive/Colab Notebooks/emisiones-2017.csv', sep = ';')
emisiones_2018 = pd.read_csv('/content/gdrive/MyDrive/Colab Notebooks/emisiones-2018 1.csv', sep = ';')
emisiones_2019 = pd.read_csv('/content/gdrive/MyDrive/Colab Notebooks/emisiones-2019 1.csv', sep = ';')
emisiones = pd.concat([emisiones_2016, emisiones_2017, emisiones_2018, emisiones_2019])
emisiones
```

```python
columnas = ['ESTACION', 'MAGNITUD', 'ANO', 'MES']
columnas.extend([col for col in emisiones if col.startswith('D')])
emisiones = emisiones[columnas]
emisiones
```

```python
emisiones = emisiones.melt(id_vars=['ESTACION', 'MAGNITUD', 'ANO', 'MES'], var_name='DIA', value_name='VALOR')
emisiones
```

```python
emisiones['DIA'] = emisiones.DIA.str.strip('D')
emisiones['FECHA'] = emisiones.ANO.apply(str) + '/' + emisiones.MES.apply(str) + '/' + emisiones.DIA.apply(str)
emisiones['FECHA'] = pd.to_datetime(emisiones.FECHA, format='%Y/%m/%d', infer_datetime_format=True, errors='coerce')
emisiones
```

```python
emisiones = emisiones.drop(emisiones[np.isnat(emisiones.FECHA)].index)
emisiones.sort_values(['ESTACION', 'MAGNITUD', 'FECHA'])
```

```python
print('Estaciones:', emisiones.ESTACION.unique())
print('Contaminantes:', emisiones.MAGNITUD.unique())
```

```python
def evolucion(estacion, contaminante, desde, hasta):
    return emisiones[(emisiones.ESTACION == estacion) & (emisiones.MAGNITUD == contaminante) & (emisiones.FECHA >= desde) & (emisiones.FECHA <= hasta)].sort_values('FECHA').VALOR
evolucion(56, 8, dt.datetime.strptime('2018/10/25', '%Y/%m/%d'), dt.datetime.strptime('2019/02/12', '%Y/%m/%d'))
```

```python
emisiones.groupby('MAGNITUD').VALOR.describe()
```

```python
emisiones.groupby(['ESTACION', 'MAGNITUD']).VALOR.describe()
```

```python
def resumen(estacion, contaminante):
    return emisiones[(emisiones.ESTACION == estacion) & (emisiones.MAGNITUD == contaminante)].VALOR.describe()

print('Resumen Dióxido de Nitrógeno en Plaza Elíptica:\n', resumen(56, 8),'\n', sep='')
print('Resumen Dióxido de Nitrógeno en Plaza del Carmen:\n', resumen(35, 8), sep='')
```

```python
def evolucion_mensual(contaminante, año):
    return emisiones[(emisiones.MAGNITUD == contaminante) & (emisiones.ANO == año)].groupby(['ESTACION', 'MES']).VALOR.mean().unstack('MES')

evolucion_mensual(8, 2019)
```
